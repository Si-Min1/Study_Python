### numpy 명령어과 컴프리헨션

```python
np.zeros((3,4)) # x[0][0] = 0, x[0][1] = 0, x[0][2] = 0, x[0][3] = 0, x[0][4] = 0 1차 배열 올려가며 이거 4번


# w*r 합산이 e값과 최대한 차이나지 않게하는 r값 찾기
w = np.array([[55,30,9],[62,22,6]]) # [몸무게, 나이, 수면시간]
e = np.array([100,150])             # 1번 혈압 2번 혈압
r = np.array([3,5,8])               # 몸무게의 r값 3, 나이의 r값 5, 수면시간의 r값 8           
e_hat = []
e1 = w*r # [55 * 3, 30 * 5, 9 * 8], [62 * 3 ...]


for i in e1:
    e_hat.append(i.sum())

e_hat = [i.sum() for i in e1]   # 컴프리헨션 위의 for문과 하는 역할이 같다



e_hat
# np.array(e_hat) - e #그냥해도 빼지기는 하는데 타입은 언제나 확인 행렬은 행렬끼리




# w.sum()           # 행렬구조가 아닌 리스트라서 실행 안됨
# np.array(w).sum() # 이렇게 바꿔주고 써야함
# np.sum(w)   # 윗줄과 역할은 같음

```
